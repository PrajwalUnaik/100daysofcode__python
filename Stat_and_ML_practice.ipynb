```python
#imported libraries 
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Sample dataset from CHATGPT.
data = np.array([
    # [Product ID, Category, Units Sold, Price per Unit, Rating]
    [101, 1, 25, 10.5, 4.2],   # Product in Category 1
    [102, 2, 40, 15.0, 3.8],   # Product in Category 2
    [103, 1, 10, 8.0, 4.5],    # Product in Category 1
    [104, 3, 5, 20.0, 4.0],    # Product in Category 3
    [105, 2, 60, 12.5, 4.1],   # Product in Category 2
    [106, 3, 30, 18.0, 3.9],   # Product in Category 3
    [107, 1, 45, 9.0, 4.7],    # Product in Category 1
    [108, 2, 20, 14.5, 3.6]    # Product in Category 2
])
```

```python

# Extract data
product_ids = data[:, 0] # Product extracted using [:, 0] whioch gives only the product
units_sold = data[:, 2] #unitsold extracted using [:, 2] which gives only the units sold

# Plot (plotting the product id extracted to the units sold )

plt.bar(product_ids, units_sold, color='red')
plt.xlabel('Product ID')
plt.ylabel('Units Sold')
plt.title('Units Sold by Product')
plt.show()
```

```python
# Extract data
categories = data[:, 1]
unique_categories, category_counts = np.unique(categories, return_counts=True)

# Plot
plt.pie(category_counts, labels=unique_categories, autopct='%1.1f%%', colors=['gold', 'lightcoral', 'skyblue'])
plt.title('Units Sold by Category')
plt.show()
```
