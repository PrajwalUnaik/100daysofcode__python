```python
import pandas as pd
```

```python
import numpy as np
from scipy import stats
```

```python
data = [12, 15, 14, 10, 18, 20, 22, 17, 19, 13, 16, 15, 14, 21, 23]
```

```python
mean = np.mean(data)
mean
```

```python
median_value = np.median(data)
median_value
```

```python
mode_value = np.argmax(np.bincount(data))
mode_value
```

```python

variance = np.var(data)
variance
```

```python
standard_deviation = np.std(data)
standard_deviation
```

```python
range = np.max(data) - np.min(data)
range
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.hist(data, edgecolor='black', bins=5)
plt.xlabel('Value', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.title('Histogram with Statistical Reference Lines', fontsize=16)


negative_standard_deviation = mean - standard_deviation
positive_standard_deviation = mean + standard_deviation

plt.axvline(variance, color='red', linestyle='-', linewidth=2, label=f'Variance: {variance:.2f}')
plt.axvline(mean, color='red', linestyle='--', linewidth=2, label=f'Mean: {mean:.2f}')
plt.axvline(negative_standard_deviation, color='red', linestyle='--', linewidth=2, label=f'Standard Deviation: {negative_standard_deviation:.2f}')
plt.axvline(positive_standard_deviation, color='red', linestyle='--', linewidth=2)
plt.axvline(median_value, color='green', linestyle=':', linewidth=2, label=f'Median: {median_value:.2f}')
plt.axvline(mode_value, color='blue', linestyle='-.', linewidth=2, label=f'Mode: {mode_value:.2f}')
plt.axvline(range, color='orange', linestyle=(0, (5, 10)), linewidth=2, label=f'Range: {range:.2f}')


plt.legend(fontsize=8, loc='upper left')

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)


plt.tight_layout()
plt.show()
```
